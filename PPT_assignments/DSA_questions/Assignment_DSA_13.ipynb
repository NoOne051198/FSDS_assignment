<font size = "3">__1. Given two linked list of the same size, the task is to create a new linked list using those linked lists. The condition is that the greater node among both linked list will be added to the new linked list.__</font>

**Examples:**

>Input: list1 = 5->2->3->8<br>
list2 = 1->7->4->5<br>
Output: New list = 5->7->4->8

>Input:list1 = 2->8->9->3<br>
list2 = 5->3->6->4<br>
Output: New list = 5->8->9->4

```python
# Define the class and function:
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next

def merge_linked_lists(list1, list2):
    if not list1 or not list2:
        return None

    new_head = None
    new_tail = None

    while list1 and list2:
        if list1.val >= list2.val:
            new_node = ListNode(list1.val)
            list1 = list1.next
        else:
            new_node = ListNode(list2.val)
            list2 = list2.next

        if not new_head:
            new_head = new_node
            new_tail = new_node
        else:
            new_tail.next = new_node
            new_tail = new_node

    # Append any remaining nodes from list1 or list2
    while list1:
        new_node = ListNode(list1.val)
        list1 = list1.next

        if not new_head:
            new_head = new_node
            new_tail = new_node
        else:
            new_tail.next = new_node
            new_tail = new_node

    while list2:
        new_node = ListNode(list2.val)
        list2 = list2.next

        if not new_head:
            new_head = new_node
            new_tail = new_node
        else:
            new_tail.next = new_node
            new_tail = new_node

    return new_head

# Helper function to print the linked list
def print_linked_list(head):
    current = head
    while current:
        print(current.val, end=" ")
        current = current.next
    print()

    
    
# Test the algorithm
# Example 1
list1_head = ListNode(5)
list1_head.next = ListNode(2)
list1_head.next.next = ListNode(3)
list1_head.next.next.next = ListNode(8)

list2_head = ListNode(1)
list2_head.next = ListNode(7)
list2_head.next.next = ListNode(4)
list2_head.next.next.next = ListNode(5)

new_list_head = merge_linked_lists(list1_head, list2_head)
print_linked_list(new_list_head)  # Output: 5 7 4 8

# Example 2
list1_head = ListNode(2)
list1_head.next = ListNode(8)
list1_head.next.next = ListNode(9)
list1_head.next.next.next = ListNode(3)

list2_head = ListNode(5)
list2_head.next = ListNode(3)
list2_head.next.next = ListNode(6)
list2_head.next.next.next = ListNode(4)

new_list_head = merge_linked_lists(list1_head, list2_head)
print_linked_list(new_list_head)  # Output: 5 8 9 4
```

<font size = "3">__2. Write a function that takes a list sorted in non-decreasing order and deletes any duplicate nodes from the list. The list should only be traversed once.__</font>

For example if the linked list is 11->11->11->21->43->43->60 then removeDuplicates() should convert the list to 11->21->43->60.

**Example 1:**

>Input:<br>
LinkedList: <br>
11->11->11->21->43->43->60<br>
Output:<br>
11->21->43->60

**Example 2:**

>Input:<br>
LinkedList: <br>
10->12->12->25->25->25->34<br>
Output:<br>
10->12->25->34

```python
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next

def remove_duplicates(head):
    if not head or not head.next:
        return head

    current = head
    while current.next:
        if current.val == current.next.val:
            current.next = current.next.next
        else:
            current = current.next

    return head

# Helper function to print the linked list
def print_linked_list(head):
    current = head
    while current:
        print(current.val, end=" ")
        current = current.next
    print()

# Test the algorithm
# Example 1
list1_head = ListNode(11)
list1_head.next = ListNode(11)
list1_head.next.next = ListNode(11)
list1_head.next.next.next = ListNode(21)
list1_head.next.next.next.next = ListNode(43)
list1_head.next.next.next.next.next = ListNode(43)
list1_head.next.next.next.next.next.next = ListNode(60)

remove_duplicates(list1_head)
print_linked_list(list1_head)  # Output: 11 21 43 60

# Example 2
list2_head = ListNode(10)
list2_head.next = ListNode(12)
list2_head.next.next = ListNode(12)
list2_head.next.next.next = ListNode(25)
list2_head.next.next.next.next = ListNode(25)
list2_head.next.next.next.next.next = ListNode(25)
list2_head.next.next.next.next.next.next = ListNode(34)

remove_duplicates(list2_head)
print_linked_list(list2_head)  # Output: 10 12 25 34
```

<font size = "3">__3. Given a linked list of size **N**. The task is to reverse every **k** nodes (where k is an input to the function) in the linked list. If the number of nodes is not a multiple of *k* then left-out nodes, in the end, should be considered as a group and must be reversed (See Example 2 for clarification).__</font>

**Example 1:**

>Input:<br>
LinkedList: 1->2->2->4->5->6->7->8<br>
K = 4<br>
Output:4 2 2 1 8 7 6 5

**Explanation:**
The first 4 elements 1,2,2,4 are reversed first
and then the next 4 elements 5,6,7,8. Hence, the
resultant linked list is 4->2->2->1->8->7->6->5.

**Example 2:**

>Input:<br>
LinkedList: 1->2->3->4->5<br>
K = 3<br>
Output:3 2 1 5 4

**Explanation:**
The first 3 elements are 1,2,3 are reversed
first and then elements 4,5 are reversed.Hence,
the resultant linked list is 3->2->1->5->4.


```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

def reverseKNodes(head, k):
    if not head or k <= 1:
        return head

    current = head
    next_head = None
    prev_tail = None

    # Count the number of nodes in the remaining list
    count = 0
    temp = head
    while temp:
        count += 1
        temp = temp.next

    # Reverse k nodes at a time
    while count >= k:
        prev = None
        current = head

        for _ in range(k):
            next_node = current.next
            current.next = prev
            prev = current
            current = next_node

        if not next_head:
            next_head = prev

        if prev_tail:
            prev_tail.next = prev

        prev_tail = head
        head = current
        count -= k

    return next_head if next_head else head

# Function to print the linked list
def printLinkedList(head):
    current = head
    while current:
        print(current.data, end=" -> ")
        current = current.next
    print("None")

# Test the algorithm
head = Node(1)
head.next = Node(2)
head.next.next = Node(2)
head.next.next.next = Node(4)
head.next.next.next.next = Node(5)
head.next.next.next.next.next = Node(6)
head.next.next.next.next.next.next = Node(7)
head.next.next.next.next.next.next.next = Node(8)

print("Original Linked list:")
printLinkedList(head)

k = 4
new_head = reverseKNodes(head, k)

print(f"Linked list after reversing every {k} nodes:")
printLinkedList(new_head)
```

<font size = "3">__4. Given a linked list, write a function to reverse every alternate k nodes (where k is an input to the function) in an efficient way. Give the complexity of your algorithm.__</font>

**Example:**

>Inputs:   1->2->3->4->5->6->7->8->9->NULL and k = 3<br>
Output:   3->2->1->4->5->6->9->8->7->NULL.

```python
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next

def reverse_alternate_k_nodes(head, k):
    def reverse_list(start, end):
        prev, curr = None, start
        while curr != end:
            next_node = curr.next
            curr.next = prev
            prev = curr
            curr = next_node
        return prev

    dummy = ListNode(0)
    dummy.next = head
    prev_group_end = dummy
    current = head
    is_alternate_group = True

    while current:
        start = current
        end = start
        # Find the end of the current group of k nodes
        for i in range(k):
            if not end:
                break
            end = end.next

        # Reverse the k nodes if this is an alternate group
        if is_alternate_group:
            prev_group_end.next = reverse_list(start, end)
            start.next = end

        # Move prev_group_end to the end of the current group
        prev_group_end = start

        # Move current to the start of the next group
        current = end
        is_alternate_group = not is_alternate_group

    return dummy.next

# Helper function to print the linked list
def print_linked_list(head):
    current = head
    while current:
        print(current.val, end=" ")
        current = current.next
    print()

# Test the algorithm
# Example
list_head = ListNode(1)
list_head.next = ListNode(2)
list_head.next.next = ListNode(3)
list_head.next.next.next = ListNode(4)
list_head.next.next.next.next = ListNode(5)
list_head.next.next.next.next.next = ListNode(6)
list_head.next.next.next.next.next.next = ListNode(7)
list_head.next.next.next.next.next.next.next = ListNode(8)
list_head.next.next.next.next.next.next.next.next = ListNode(9)

k = 3
reversed_head = reverse_alternate_k_nodes(list_head, k)
print_linked_list(reversed_head)  # Output: 3 2 1 4 5 6 9 8 7
```

<font size = "3">__5. Given a linked list and a key to be deleted. Delete last occurrence of key from linked. The list may have duplicates.__</font>

**Examples**:
>Input:   1->2->3->5->2->10, key = 2<br>
Output:  1->2->3->5->10

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

def deleteLastOccurrence(head, key):
    if not head:
        return None

    # Initialize pointers
    last_occurrence = None
    current = head
    prev = None

    # Traverse the linked list
    while current:
        if current.data == key:
            last_occurrence = prev

        prev = current
        current = current.next

    # If last_occurrence is None, key not found
    if not last_occurrence:
        return head

    # Update the next pointer of the previous node
    last_occurrence.next = last_occurrence.next.next

    return head

# Function to print the linked list
def printLinkedList(head):
    current = head
    while current:
        print(current.data, end=" -> ")
        current = current.next
    print("None")

# Test the algorithm
head = Node(1)
head.next = Node(2)
head.next.next = Node(3)
head.next.next.next = Node(5)
head.next.next.next.next = Node(2)
head.next.next.next.next.next = Node(10)

print("Original Linked list:")
printLinkedList(head)

key = 2
new_head = deleteLastOccurrence(head, key)

print(f"Linked list after deleting last occurrence of {key}:")
printLinkedList(new_head)
```

<font size = "3">__6. Given two sorted linked lists consisting of **N** and **M** nodes respectively. The task is to merge both of the lists (in place) and return the head of the merged list.__</font>

**Examples:**

>Input: a: 5->10->15, b: 2->3->20<br>
Output: 2->3->5->10->15->20

>Input: a: 1->1, b: 2->4<br>
Output: 1->1->2->4

```python
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next

def merge_sorted_lists(a, b):
    # Create a dummy node to serve as the new head
    dummy = ListNode()
    current = dummy

    # Traverse both lists and attach the smaller node to the merged list
    while a and b:
        if a.val <= b.val:
            current.next = a
            a = a.next
        else:
            current.next = b
            b = b.next
        current = current.next

    # Attach the remaining nodes from list a or list b (if any)
    if a:
        current.next = a
    else:
        current.next = b

    return dummy.next

# Helper function to print the linked list
def print_linked_list(head):
    current = head
    while current:
        print(current.val, end=" ")
        current = current.next
    print()

# Test the algorithm
# Example 1
list_a = ListNode(5)
list_a.next = ListNode(10)
list_a.next.next = ListNode(15)

list_b = ListNode(2)
list_b.next = ListNode(3)
list_b.next.next = ListNode(20)

merged_head = merge_sorted_lists(list_a, list_b)
print_linked_list(merged_head)  # Output: 2 3 5 10 15 20

# Example 2
list_a = ListNode(1)
list_a.next = ListNode(1)

list_b = ListNode(2)
list_b.next = ListNode(4)

merged_head = merge_sorted_lists(list_a, list_b)
print_linked_list(merged_head)  # Output: 1 1 2 4
```

<font size = "3">__7. Given a **Doubly Linked List**, the task is to reverse the given Doubly Linked List.__</font>

**Example:**
>Original Linked list 10 8 4 2<br>
Reversed Linked list 2 4 8 10

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.prev = None
        self.next = None

def reverseDoublyLinkedList(head):
    if not head:
        return None

    prev = None
    current = head

    # Swap prev and next pointers for all nodes
    while current:
        temp = current.next
        current.next = prev
        current.prev = temp
        prev = current
        current = temp

    return prev

# Function to print the doubly linked list
def printDoublyLinkedList(head):
    current = head
    while current:
        print(current.data, end=" ")
        current = current.next
    print()

# Test the algorithm
head = Node(10)
head.next = Node(8)
head.next.prev = head
head.next.next = Node(4)
head.next.next.prev = head.next
head.next.next.next = Node(2)
head.next.next.next.prev = head.next.next

print("Original Linked list:", end=" ")
printDoublyLinkedList(head)

reversed_head = reverseDoublyLinkedList(head)

print("Reversed Linked list:", end=" ")
printDoublyLinkedList(reversed_head)
```

<font size = "3">__8. Given a doubly linked list and a position. The task is to delete a node from given position in a doubly linked list.__</font>

**Example 1:**

>Input:<br>
LinkedList = 1 <--> 3 <--> 4<br>
x = 3<br>
Output:1 3

**Explanation:** After deleting the node at
position 3 (position starts from 1),
the linked list will be now as 1->3.

**Example 2:**

>Input:<br>
LinkedList = 1 <--> 5 <--> 2 <--> 9<br>
x = 1<br>
Output:5 2 9

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None
        self.prev = None

class DoublyLinkedList:
    def __init__(self):
        self.head = None

    def append(self, data):
        new_node = Node(data)
        if not self.head:
            self.head = new_node
        else:
            current = self.head
            while current.next:
                current = current.next
            current.next = new_node
            new_node.prev = current

    def delete_at_position(self, position):
        if position == 1:
            self.head = self.head.next
            self.head.prev = None
        else:
            current = self.head
            count = 1
            while current:
                if count == position:
                    prev_node = current.prev
                    next_node = current.next
                    prev_node.next = next_node
                    if next_node:
                        next_node.prev = prev_node
                    break
                count += 1
                current = current.next

    def display(self):
        current = self.head
        while current:
            print(current.data, end=" ")
            current = current.next
        print()

# Test the algorithm
dll = DoublyLinkedList()
dll.append(1)
dll.append(3)
dll.append(4)

print("Original Linked list:")
dll.display()  # Output: 1 3 4

position = 3
dll.delete_at_position(position)

print(f"Linked list after deleting node at position {position}:")
dll.display()  # Output: 1 3
```
