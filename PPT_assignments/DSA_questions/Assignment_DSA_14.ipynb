<font size="3">__1. Given a linked list of **N** nodes such that it may contain a loop.__</font>

A loop here means that the last node of the link list is connected to the node at position X(1-based index). If the link list does not have any loop, X=0.

Remove the loop from the linked list, if it is present, i.e. unlink the last node which is forming the loop.

**Example 1:**

>Input:<br>
N = 3<br>
value[] = {1,3,4}<br>
X = 2<br>
Output:1<br>
Explanation:The link list looks like<br>

     ^    |<br>
     |____|<br>

**Example 2:**
>Input:<br>

**Example 3:**

>Input:<br>
N = 4<br>
value[] = {1,2,3,4}<br>
X = 1<br><br>
Output:1<br>
Explanation:The link list looks like<br>
1 -> 2 -> 3 -> 4<br>
^              |<br>
|______________|<br>
A loop is present.<br>
If you remove it successfully,<br>
the answer will be 1.<br>

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

class LinkedList:
    def __init__(self):
        self.head = None

    def append(self, data):
        new_node = Node(data)
        if not self.head:
            self.head = new_node
        else:
            current = self.head
            while current.next:
                current = current.next
            current.next = new_node

    def detect_and_remove_loop(self):
        slow_ptr = self.head
        fast_ptr = self.head
        loop_exists = False

        # Detect loop using Floyd's cycle-finding algorithm
        while slow_ptr and fast_ptr and fast_ptr.next:
            slow_ptr = slow_ptr.next
            fast_ptr = fast_ptr.next.next
            if slow_ptr == fast_ptr:
                loop_exists = True
                break

        if loop_exists:
            slow_ptr = self.head
            while slow_ptr.next != fast_ptr.next:
                slow_ptr = slow_ptr.next
                fast_ptr = fast_ptr.next

            fast_ptr.next = None  # Remove the loop

    def display(self):
        current = self.head
        while current:
            print(current.data, end=" ")
            current = current.next
        print()

# Test the algorithm
ll = LinkedList()
ll.append(1)
ll.append(3)
ll.append(4)

# Create a loop for testing
ll.head.next.next.next = ll.head.next

ll.detect_and_remove_loop()
ll.display()  # Output: 1 3 4
```

<font size="3">__2. A number **N** is represented in Linked List such that each digit corresponds to a node in linked list. You need to add 1 to it.__</font>

**Example 1:**

>Input:<br>
LinkedList: 4->5->6<br>
Output:457


**Example 2:**

>Input:<br>
LinkedList: 1->2->3<br>
Output:124

```python
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next

def add_one_to_linked_list(head):
    # Helper function to reverse the linked list
    def reverse_linked_list(node):
        prev = None
        current = node
        while current:
            next_node = current.next
            current.next = prev
            prev = current
            current = next_node
        return prev

    # Reverse the linked list to add 1
    head = reverse_linked_list(head)

    # Add 1 to the linked list
    carry = 1
    current = head
    while current:
        sum_val = current.val + carry
        carry = sum_val // 10
        current.val = sum_val % 10
        prev = current
        current = current.next

    # If carry is not 0, add a new node with carry
    if carry:
        prev.next = ListNode(carry)

    # Reverse the linked list back to its original form
    head = reverse_linked_list(head)
    return head

def print_linked_list(head):
    current = head
    while current:
        print(current.val, end=" -> ")
        current = current.next
    print("None")

# Test the algorithm
# Example 1
head1 = ListNode(4)
head1.next = ListNode(5)
head1.next.next = ListNode(6)
result1 = add_one_to_linked_list(head1)
print_linked_list(result1)  # Output: 4 -> 5 -> 7 -> None

# Example 2
head2 = ListNode(1)
head2.next = ListNode(2)
head2.next.next = ListNode(3)
result2 = add_one_to_linked_list(head2)
print_linked_list(result2)  # Output: 1 -> 2 -> 4 -> None
```

<font size="3">__3. Given a Linked List of size N, where every node represents a sub-linked-list and contains two pointers:__</font><br>(i) a **next** pointer to the next node,(ii) a **bottom** pointer to a linked list where this node is head.Each of the sub-linked-list is in sorted order.Flatten the Link List such that all the nodes appear in a single level while maintaining the sorted order. **Note:** The flattened list will be printed using the bottom pointer instead of next pointer.

**Example 1:**

>Input:<br>
5 -> 10 -> 19 -> 28<br>
|     |     |     |<br>
7     20    22   35
|           |     |<br>
8          50    40<br>
|                 |<br>
30               45<br>
Output: 5-> 7-> 8- > 10 -> 19-> 20-><br>
22-> 28-> 30-> 35-> 40-> 45-> 50.<br>
Explanation:<br>
The resultant linked lists has every
node in a single level.(Note:| represents the bottom pointer.)

**Example 2:**

>Input:<br>
5 -> 10 -> 19 -> 28<br>
|          |<br>
7          22<br>
|          |<br>
8          50<br>
|<br>
30<br>
Output: 5->7->8->10->19->22->28->30->50<br>
Explanation:<br>
The resultant linked lists has every
node in a single level.

`(Note:| represents the bottom pointer.)`

```python
class ListNode:
    def __init__(self, val=0, next=None, bottom=None):
        self.val = val
        self.next = next
        self.bottom = bottom

def merge_sorted_lists(a, b):
    if not a:
        return b
    if not b:
        return a

    result = None
    if a.val < b.val:
        result = a
        result.bottom = merge_sorted_lists(a.bottom, b)
    else:
        result = b
        result.bottom = merge_sorted_lists(a, b.bottom)

    result.next = None
    return result

def flatten_linked_list(head):
    if not head or not head.next:
        return head

    # Merge the current list with the flattened next list
    head.next = flatten_linked_list(head.next)

    # Merge the current list with the flattened bottom list
    head = merge_sorted_lists(head, head.next)

    return head

def print_linked_list(head):
    current = head
    while current:
        print(current.val, end=" -> ")
        current = current.bottom
    print("None")

# Test the algorithm
# Example 1
head1 = ListNode(5, ListNode(10, ListNode(19, ListNode(28))))
head1.bottom = ListNode(7, ListNode(20, ListNode(22, ListNode(35))))
head1.bottom.bottom = ListNode(8, ListNode(50))
head1.bottom.bottom.bottom = ListNode(30)
result1 = flatten_linked_list(head1)
print_linked_list(result1)  # Output: 5 -> 7 -> 8 -> 10 -> 19 -> 20 -> 22 -> 28 -> 30 -> 35 -> 50 -> None

# Example 2
head2 = ListNode(5, ListNode(10, ListNode(19, ListNode(28))))
head2.bottom = ListNode(7)
head2.bottom.bottom = ListNode(8, ListNode(50))
head2.bottom.bottom.bottom = ListNode(30)
result2 = flatten_linked_list(head2)
print_linked_list(result2)  # Output: 5 -> 7 -> 8 -> 10 -> 19 -> 28 -> 30 -> 50 -> None
```

<font size="3">__4. You are given a special linked list with N nodes where each node has a next pointer pointing to its next node. You are also given M random pointers, where you will be given **M** number of pairs denoting two nodes **a** and **b**  **i.e. a->arb = b** (arb is pointer to random node).__</font><br>

Construct a copy of the given list. The copy should consist of exactly **N** new nodes, where each new node has its value set to the value of its corresponding original node. Both the next and random pointer of the new nodes should point to new nodes in the copied list such that the pointers in the original list and copied list represent the same list state. None of the pointers in the new list should point to nodes in the original list.

For example, if there are two nodes **X** and **Y** in the original list, where **X.arb** **-->** **Y**, then for the corresponding two nodes **x** and **y** in the copied list, **x.arb --> y.**

Return the head of the copied linked list.

**Note** :- The diagram isn't part of any example, it just depicts an example of how the linked list may look like.

**Example 1:**
>Input:<br>
N = 4, M = 2<br>
value = {1,2,3,4}<br>
pairs = {{1,2},{2,4}}<br>
Output:1<br>
Explanation:In this test case, there
are 4 nodes in linked list.  Among these
4 nodes,  2 nodes have arbitrary pointer
set, rest two nodes have arbitrary pointer
as NULL. Second line tells us the value
of four nodes. The third line gives the
information about arbitrary pointers.
The first node arbitrary pointer is set to

**Example 2:**
>Input:<br>
N = 4, M = 2<br>
value[] = {1,3,5,9}<br>
pairs[] = {{1,1},{3,4}}<br>
Output:1<br>
Explanation:In the given testcase ,
applying the method as stated in the
above example, the output will be 1.


```python
class ListNode:
    def __init__(self, val=0, next=None, random=None):
        self.val = val
        self.next = next
        self.random = random

def copyRandomList(head):
    if not head:
        return None

    # Step 1: Create new nodes and insert them between the original nodes
    current = head
    while current:
        new_node = ListNode(current.val)
        new_node.next = current.next
        current.next = new_node
        current = new_node.next

    # Step 2: Update random pointers of the new nodes
    current = head
    while current:
        if current.random:
            current.next.random = current.random.next
        current = current.next.next

    # Step 3: Separate the original and copied lists
    new_head = head.next
    current = head
    while current:
        new_node = current.next
        current.next = new_node.next
        if new_node.next:
            new_node.next = new_node.next.next
        current = current.next

    return new_head

# Helper function to print the linked list
def print_linked_list(head):
    current = head
    while current:
        print(f"({current.val}, {current.random.val if current.random else None})", end=" -> ")
        current = current.next
    print("None")

# Test the algorithm
# Example 1
head1 = ListNode(1)
node2 = ListNode(2)
node3 = ListNode(3)
node4 = ListNode(4)
head1.next = node2
node2.next = node3
node3.next = node4
head1.random = node2
node2.random = node4
result1 = copyRandomList(head1)
print_linked_list(result1)  # Output: (1, 2) -> (2, 4) -> (3, None) -> (4, None) -> None

# Example 2
head2 = ListNode(1)
node3 = ListNode(3)
node5 = ListNode(5)
node9 = ListNode(9)
head2.next = node3
node3.next = node5
node5.next = node9
head2.random = head2
node3.random = node9
result2 = copyRandomList(head2)
print_linked_list(result2)  # Output: (1, 1) -> (3, 9) -> (5, None) -> (9, None) -> None
```

<font size="3">__5. Given the head of a singly linked list, group all the nodes with odd indices together followed by the nodes with even indices, and return *the reordered list*.__</font><br>

The **first** node is considered **odd**, and the **second** node is **even**, and so on.

Note that the relative order inside both the even and odd groups should remain as it was in the input.

You must solve the problem in `O(1)` extra space complexity and `O(n)` time complexity.

**Example 1:**

>Input: head = [1,2,3,4,5]<br>
Output: [1,3,5,2,4]

**Example 2:**

>Input: head = [2,1,3,5,6,4,7]<br>
Output: [2,3,6,7,1,5,4]

```python
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next

def oddEvenList(head):
    if not head:
        return None

    odd_head = head
    even_head = head.next
    odd_tail = odd_head
    even_tail = even_head

    while even_tail and even_tail.next:
        odd_tail.next = even_tail.next
        odd_tail = odd_tail.next
        even_tail.next = odd_tail.next
        even_tail = even_tail.next

    odd_tail.next = even_head
    return odd_head

# Helper function to print the linked list
def print_linked_list(head):
    current = head
    while current:
        print(current.val, end=" -> ")
        current = current.next
    print("None")

# Test the algorithm
# Example 1
head1 = ListNode(1)
node2 = ListNode(2)
node3 = ListNode(3)
node4 = ListNode(4)
node5 = ListNode(5)
head1.next = node2
node2.next = node3
node3.next = node4
node4.next = node5
result1 = oddEvenList(head1)
print_linked_list(result1)  # Output: 1 -> 3 -> 5 -> 2 -> 4 -> None

# Example 2
head2 = ListNode(2)
node1 = ListNode(1)
node3 = ListNode(3)
node5 = ListNode(5)
node6 = ListNode(6)
node4 = ListNode(4)
node7 = ListNode(7)
head2.next = node1
node1.next = node3
node3.next = node5
node5.next = node6
node6.next = node4
node4.next = node7
result2 = oddEvenList(head2)
print_linked_list(result2)  # Output: 2 -> 3 -> 6 -> 7 -> 1 -> 5 -> 4 -> None
```

<font size="3">__6. Given a singly linked list of size **N**. The task is to **left-shift** the linked list by **k** nodes, where **k** is a given positive integer smaller than or equal to length of the linked list.__</font>

**Example 1:**

>Input:<br>
N = 5<br>
value[] = {2, 4, 7, 8, 9}<br>
k = 3<br>
Output:8 9 2 4 7<br>
Explanation:Rotate 1:4 -> 7 -> 8 -> 9 -> 2<br>
Rotate 2: 7 -> 8 -> 9 -> 2 -> 4<br>
Rotate 3: 8 -> 9 -> 2 -> 4 -> 7

**Example 2:**

>Input:<br>
N = 8<br>
value[] = {1, 2, 3, 4, 5, 6, 7, 8}<br>
k = 4<br>
Output:5 6 7 8 1 2 3 4

```python
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next

def rotateLeft(head, k):
    if not head or k == 0:
        return head

    # Find the length of the linked list
    length = 0
    current = head
    while current:
        length += 1
        current = current.next

    # Handle cases where k is greater than N
    k = k % length
    if k == 0:
        return head

    # Find the new head and new tail of the rotated list
    new_tail_position = length - k - 1
    current = head
    for _ in range(new_tail_position):
        current = current.next

    new_head = current.next
    current.next = None

    # Traverse to the old tail of the list and connect it to the old head
    current = new_head
    while current.next:
        current = current.next
    current.next = head

    return new_head

# Helper function to print the linked list
def print_linked_list(head):
    current = head
    while current:
        print(current.val, end=" -> ")
        current = current.next
    print("None")

# Test the algorithm
# Example 1
head1 = ListNode(2)
node2 = ListNode(4)
node3 = ListNode(7)
node4 = ListNode(8)
node5 = ListNode(9)
head1.next = node2
node2.next = node3
node3.next = node4
node4.next = node5
k1 = 3
result1 = rotateLeft(head1, k1)
print_linked_list(result1)  # Output: 8 -> 9 -> 2 -> 4 -> 7 -> None

# Example 2
head2 = ListNode(1)
node2 = ListNode(2)
node3 = ListNode(3)
node4 = ListNode(4)
node5 = ListNode(5)
node6 = ListNode(6)
node7 = ListNode(7)
node8 = ListNode(8)
head2.next = node2
node2.next = node3
node3.next = node4
node4.next = node5
node5.next = node6
node6.next = node7
node7.next = node8
k2 = 4
result2 = rotateLeft(head2, k2)
print_linked_list(result2)  # Output: 5 -> 6 -> 7 -> 8 -> 1 -> 2 -> 3 -> 4 -> None
```

<font size="3">__7. You are given the head of a linked list with n nodes.__</font><br>

For each node in the list, find the value of the **next greater node**. That is, for each node, find the value of the first node that is next to it and has a **strictly larger** value than it.

Return an integer array `answer` where `answer[i]` is the value of the next greater node of the `ith` node (**1-indexed**). If the `ith` node does not have a next greater node, set `answer[i] = 0`.

**Example 1:**

>Input: head = [2,1,5]<br>
Output: [5,5,0]

**Example 2:**

>Input: head = [2,7,4,3,5]<br>
Output: [7,0,5,5,0]

```python
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next

def nextGreaterNodes(head):
    # Step 1: Convert the linked list to an array
    arr = []
    current = head
    while current:
        arr.append(current.val)
        current = current.next

    n = len(arr)
    answer = [0] * n
    stack = []  # To keep track of indices with potential next greater nodes

    # Step 3: Traverse the array from right to left
    for i in range(n - 1, -1, -1):
        # Step 4: Pop nodes from the stack until finding a node with a greater value
        while stack and arr[i] >= arr[stack[-1]]:
            stack.pop()

        # Step 5: If the stack is empty, there is no next greater node
        # for the current node, so the value will be 0
        answer[i] = 0 if not stack else arr[stack[-1]]

        # Step 6: Update the answer array and push the current index to the stack
        stack.append(i)

    return answer

# Helper function to create a linked list from a list
def create_linked_list(arr):
    dummy = ListNode()
    current = dummy
    for val in arr:
        current.next = ListNode(val)
        current = current.next
    return dummy.next

# Test the algorithm
# Example 1
head1 = create_linked_list([2, 1, 5])
result1 = nextGreaterNodes(head1)
print(result1)  # Output: [5, 5, 0]

# Example 2
head2 = create_linked_list([2, 7, 4, 3, 5])
result2 = nextGreaterNodes(head2)
print(result2)  # Output: [7, 0, 5, 5, 0]
```

<font size="3">__8. Given the head of a linked list, we repeatedly delete consecutive sequences of nodes that sum to 0 until there are no such sequences.__</font>

After doing so, return the head of the final linked list.  You may return any such answer.

(Note that in the examples below, all sequences are serializations of `ListNode` objects.)

**Example 1:**

>Input: head = [1,2,-3,3,1]<br>
Output: [3,1]<br>
Note: The answer [1,2,1] would also be accepted.

**Example 2:**

>Input: head = [1,2,3,-3,4]<br>
Output: [1,2,4]


**Example 3:**

>Input: head = [1,2,3,-3,-2]<br>
Output: [1]

```python
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next

def removeZeroSumSublists(head):
    dummy = ListNode(0)
    dummy.next = head
    current = dummy
    current_sum = 0
    sum_dict = {}

    while current:
        current_sum += current.val

        if current_sum in sum_dict:
            start = sum_dict[current_sum].next
            temp_sum = current_sum
            while start != current:
                temp_sum += start.val
                del sum_dict[temp_sum]
                start = start.next

            sum_dict[current_sum].next = current.next

        else:
            sum_dict[current_sum] = current

        current = current.next

    return dummy.next

# Helper function to create a linked list from a list
def create_linked_list(arr):
    dummy = ListNode()
    current = dummy
    for val in arr:
        current.next = ListNode(val)
        current = current.next
    return dummy.next

# Helper function to convert a linked list to a list
def linked_list_to_list(head):
    result = []
    current = head
    while current:
        result.append(current.val)
        current = current.next
    return result

# Test the algorithm
# Example 1
head1 = create_linked_list([1, 2, -3, 3, 1])
result1 = removeZeroSumSublists(head1)
print(linked_list_to_list(result1))  # Output: [3, 1]

# Example 2
head2 = create_linked_list([1, 2, 3, -3, 4])
result2 = removeZeroSumSublists(head2)
print(linked_list_to_list(result2))  # Output: [1, 2, 4]

# Example 3
head3 = create_linked_list([1, 2, 3, -3, -2])
result3 = removeZeroSumSublists(head3)
print(linked_list_to_list(result3))  # Output: [1]
```
